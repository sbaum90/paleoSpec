```python
# This notebook reproduces the spectrum plots in 
# https://arxiv.org/pdf/1806.05991.pdf. Note that 
# the solar and DSNB neutrino flux spectra in 
# CalcSpectra are slightly different than those used
# in the code which produced these plots, giving
# rise to small difference in the neutrino spectra below
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import sys
sys.path.append("../../")
from paleoSpec import CalcSpectra
```

```python
# Set up some plotting settings
fs = 14
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
plt.rcParams.update({'legend.fontsize': fs,
                     'axes.labelsize': fs,
                     'axes.titlesize': fs,
                     'xtick.labelsize': fs,
                     'ytick.labelsize': fs})
plt.rcParams["figure.figsize"] = (7, 5)

plt_colors = ['k',
              '#1b9e77',
              '#d95f02',
              '#7570b3',
              '#e7298a',
              '#66a61e',
              '#e6ab02',
              '#a6761d',
              '#666666']
```

```python
# Reproduce Fig 1 from https://arxiv.org/pdf/1806.05991.pdf
mineral = "Epsomite"
Calc_Eps_wH = CalcSpectra.CalcSpectra(mineral, switch_keep_H=True)

ranges = Calc_Eps_wH.Trange
element_names = Calc_Eps_wH.TargetList.nameT_list

plt.close('all')
for i, name in enumerate(element_names):
    plt.plot(ranges[i][:,0], 0.1*ranges[i][:,2], label=name)

plt.xscale('log')
plt.yscale('log')
plt.xlim(1e-2,1e3)
plt.ylim(1e-1,1e4)
plt.xlabel(r'$E_R$ [keV]')
plt.ylabel(r'$x_T$ [nm]')
plt.legend()
plt.tick_params(right=True,top=True)
plt.tick_params(which='minor',right=True,top=True)
plt.title(mineral+"; Ranges", y=1.02)
plt.tight_layout()
plt.show()
```

```python
# Reproduce Fig 2 from https://arxiv.org/pdf/1806.05991.pdf
mineral = "Epsomite"
Calc_Eps_woH = CalcSpectra.CalcSpectra(mineral, switch_keep_H=False)

C238 = 1e-10

bkg_nu_solar = Calc_Eps_woH.calc_dRdx_BkgNeu_solar()
bkg_nu_DSNB = Calc_Eps_woH.calc_dRdx_BkgNeu_DSNB()
bkg_nu_atm = Calc_Eps_woH.calc_dRdx_BkgNeu_atm()
bkg_rad_neutrons = Calc_Eps_woH.calc_dRdx_Bkgn(C238)

sig_5 = Calc_Eps_woH.calc_dRdx_MW(5., 1e-45)
sig_500 = Calc_Eps_woH.calc_dRdx_MW(500., 1e-45)

plt.close('all')
plt.plot(0.1*bkg_nu_solar[0], 10.*(bkg_nu_solar[1]+bkg_nu_DSNB[1]+bkg_nu_atm[1]), color=plt_colors[1], linestyle=':', label=r'$\nu$')
plt.plot(0.1*bkg_rad_neutrons[0], 10.*bkg_rad_neutrons[1], color=plt_colors[2], linestyle='--', label=r'$n$')
plt.plot(0.1*sig_5[0], 10.*sig_5[1], color=plt_colors[3], linestyle='-', label=r'$5\,$GeV, $10^{-45}\,{\rm cm}^2$')
plt.plot(0.1*sig_500[0], 10.*sig_500[1], color=plt_colors[4], linestyle='-.', label=r'$500\,$GeV, $10^{-45}\,{\rm cm}^2$')

plt.xscale('log')
plt.yscale('log')
plt.xlim(1e0,1e3)
plt.ylim(1e-1,1e6)
plt.xlabel(r'$x_T$ [nm]')
plt.ylabel(r'$dR/dx_T$ [nm$^{-1}$\,kg$^{-1}$\,Myr$^{-1}$]')
plt.legend()
plt.tick_params(right=True,top=True)
plt.tick_params(which='minor',right=True,top=True)
plt.title(mineral+r"; $C^{238}/$[g/g] = "+str(C238), y=1.02)
plt.tight_layout()
plt.show()
```
